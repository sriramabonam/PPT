```python
from pyhive import hive

# Establish a connection to Hive
connection = hive.connect(host='localhost', port=10000, username='your_username')

# Create a cursor object
cursor = connection.cursor()

# Execute the HiveQL query to create the table
create_table_query = '''
    CREATE TABLE Employees (
        id INT,
        name STRING,
        salary FLOAT
    )
'''

cursor.execute(create_table_query)

# Commit the changes
connection.commit()

# Close the cursor and connection
cursor.close()
connection.close()
```

```python
from pyhive import hive

# Establish a connection to Hive
connection = hive.Connection(host='localhost', port=10000, username='your_username')

# Create a cursor object
cursor = connection.cursor()

# Execute the HiveQL query to perform the inner join
inner_join_query = '''
    SELECT *
    FROM Orders
    INNER JOIN Customers ON Orders.customer_id = Customers.customer_id
'''

cursor.execute(inner_join_query)

# Fetch all the retrieved records
results = cursor.fetchall()

# Print the retrieved records
for row in results:
    print(row)

# Close the cursor and connection
cursor.close()
connection.close()
```
