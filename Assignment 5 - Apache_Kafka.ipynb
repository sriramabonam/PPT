```python
c) Implement logic to send messages to a Kafka topic.
from confluent_kafka import Producer
import json

# Configure the Kafka broker's addresses (bootstrap servers)
bootstrap_servers = 'localhost:9092'

# Create the producer configuration
producer_config = {
    'bootstrap.servers': bootstrap_servers,
    'client.id': 'my_producer'
}

# Create the Kafka producer
producer = Producer(producer_config)

# Define the Kafka topic to which messages will be sent
topic = 'my_topic'

# Function to handle the delivery reports
def delivery_report(err, msg):
    if err is not None:
        print(f'Failed to deliver message: {err}')
    else:
        print(f'Message delivered to topic: {msg.topic()} | partition: {msg.partition()} | offset: {msg.offset()}')

# Function to send a message to Kafka
def send_message(topic, message):
    producer.produce(topic=topic, value=json.dumps(message), callback=delivery_report)
    producer.flush()

# Example messages
messages = [
    {'id': 1, 'name': 'John Doe'},
    {'id': 2, 'name': 'Jane Smith'},
    {'id': 3, 'name': 'Alice Johnson'}
]

# Send each message to Kafka
for message in messages:
    send_message(topic, message)

# Close the Kafka producer
producer.close()
```

```python
4. Producing and Consuming Messages
 a) Write a Python program to produce messages to a Kafka topic.
from confluent_kafka import Producer

# Configure the Kafka broker's addresses (bootstrap servers)
bootstrap_servers = 'localhost:9092'

# Create the producer configuration
producer_config = {
    'bootstrap.servers': bootstrap_servers,
    'client.id': 'my_producer'
}

# Create the Kafka producer
producer = Producer(producer_config)

# Define the Kafka topic to which messages will be sent
topic = 'my_topic'

# Produce messages to Kafka
messages = [
    'Hello, Kafka!',
    'This is a test message.',
    'Another message to send.'
]

for message in messages:
    # Produce the message
    producer.produce(topic=topic, value=message.encode('utf-8'))

# Flush any remaining messages in the producer
producer.flush()

# Close the Kafka producer
producer.close()
```

```python
b) Implement logic to consume messages from the same Kafka topic.
from confluent_kafka import Consumer, KafkaError

# Configure the Kafka broker's addresses (bootstrap servers)
bootstrap_servers = 'localhost:9092'

# Create the consumer configuration
consumer_config = {
    'bootstrap.servers': bootstrap_servers,
    'group.id': 'my_consumer_group',
    'auto.offset.reset': 'earliest'
}

# Create the Kafka consumer
consumer = Consumer(consumer_config)

# Subscribe to the Kafka topic
topic = 'my_topic'
consumer.subscribe([topic])

# Consume messages from Kafka
try:
    while True:
        message = consumer.poll(1.0)

        if message is None:
            continue

        if message.error():
            if message.error().code() == KafkaError._PARTITION_EOF:
                print(f'Reached end of partition {message.partition()}')
                continue
            else:
                print(f'Consumer error: {message.error()}')
                continue

        # Process the received message
        print(f'Received message: {message.value().decode("utf-8")}')

except KeyboardInterrupt:
    # Close the Kafka consumer on interrupt
    consumer.close()
```
